```python
doc = pd.read_csv('../text.csv')
```

```python
doc.iloc[:,0],doc.shape
```

```python
def parse_line(line):
    return '|'+ '|'.join(line) + '|'

def parse_header(header):
    return '|' + '|'.join(header) + '|\n' + '|' + '---|'* len(header)

def parse_csv():
    print(parse_header(doc.columns))
    for i in range(doc.shape[0]):
        line = doc.iloc[i,:]
        print(parse_line(line), end='')
        print()
```

```python
parse_csv()
```

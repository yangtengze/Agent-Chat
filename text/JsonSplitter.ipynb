```python
import orjson
```

```python
def flatten_json(obj, parent_key='', sep='.'):
    """
    将嵌套 JSON 展平为键值对
    例如: {"a": {"b": 1}} -> {"a.b": 1}
    """
    items = []
    
    if isinstance(obj, dict):
        for k, v in obj.items():
            new_key = f"{parent_key}{sep}{k}" if parent_key else k
            items.extend(flatten_json(v, new_key, sep).items())
            
    elif isinstance(obj, list):
        for i, v in enumerate(obj):
            new_key = f"{parent_key}[{i}]" if parent_key else str(i)
            # 如果列表元素是基本类型，直接合并为字符串（更紧凑）
            if isinstance(v, (str, int, float, bool)):
                items.append((parent_key, ', '.join(str(x) for x in obj)))
                break
            else:
                items.extend(flatten_json(v, new_key, sep).items())
                
    else:
        # 基本类型
        items.append((parent_key, str(obj)))
        
    return dict(items)


def parse_json_for_rag(json_path):
    """
    解析 JSON 文件为文本块
    """
    with open(json_path, 'r', encoding='utf-8') as f:
        json_bytes = f.read()
    data = orjson.loads(json_bytes)
    # 展平
    flat_data = flatten_json(data)
    
    # 转为文本行
    lines = [f"{k}: {v}" for k, v in flat_data.items()]
    text = '\n'.join(lines)
    return text
```

```python
# 使用示例
json_text = parse_json_for_rag('./text_files/text.json')
print(json_text)
```

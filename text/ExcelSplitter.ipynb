```python
import pandas as pd
```

```python
def __parse_excel_df(excel_df):
    markdown_text = []
    size = excel_df.shape[1]
    for i in range(excel_df.shape[0]):
        cells = [str(x).replace('\n', ' ').replace('\r', '') for x in excel_df.iloc[i, :]]
        text = '|' + '|'.join(cells) + '|'
        markdown_text.append(text)
        if i == 0:
            text = '|' + '---|' * size
            markdown_text.append(text)
    return '\n'.join(markdown_text)
```

```python
excel = pd.read_excel('./text_files/text.xlsx')
```

```python
markdown_text = __parse_excel_df(excel)
```
